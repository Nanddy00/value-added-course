# iv) MISING VALUES

```python
import numpy as np
import pandas as pd
```

```python
datas={'first score':[78,56,np.nan,98],
      'second score':[np.nan,76,np.nan,52],
      'third score':[70,np.nan,45,84]}
```

```python
scores=pd.DataFrame(datas)
```

```python
scores
```

```python
scores.isnull()  #to view a null value
```

```python
scores.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
impute.fit(scores)
```

```python
scores=impute.transform(scores)
```

```python
print(scores)
```
