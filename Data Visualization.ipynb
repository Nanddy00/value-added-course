```python
movies=['kaithi','vikram','master','leo','kgf','vtv','okok']
ratings=[80,49,85,99,78,63,50]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# scatter plot

```python
plt.scatter(movies,ratings)
```

```python
plt.bar(movies,ratings)
```

```python
plt.plot(movies,ratings,linestyle='dotted')  #dashdot,solid,
```

```python
plt.pie(ratings, labels=movies)
```

# working with inbuilt data base

```python
sb.get_dataset_names()
```

```python
tip=sb.load_dataset('tips')
```

```python
tip
```

```python
tip.tail() # to view a last rows
```

# Relationship plot

```python
sb.relplot(x='total_bill',y='tip',data=tip,col='time',hue="day",style="day",kind="line") #line or scatter
```

# scatter box plot

```python
sb.catplot(x='day',y='total_bill',kind='box',data=tip)
```

# REGRESSION PLOT

```python
sb.regplot(x='total_bill',y='tip',data=tip)
```

# pair plot

```python
sb.pairplot(tip,)
```
